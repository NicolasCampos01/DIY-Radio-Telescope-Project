```python
import matplotlib.pyplot as plt
import numpy as np
import os, glob
import csv
import scipy as sp
#import plot_params
%matplotlib inline
```

```python
# function to look at all the spectra in a directory (e.g., lon60) to check for interference
# if you see anything bad, delete the file manually (or rename to something like prefix.csv_bad) before proceeding
def inspect(directory):
    files = glob.glob(directory + '/*.csv')
    nfiles = len(files)
    if nfiles == 0:
        print('No files found in directory ', directory)
        return 0
    else:
        fig = plt.figure(figsize=(12, 8))
        ax = fig.add_subplot(111)

        for n, f in enumerate(files):
            nu1, C1 = np.genfromtxt(f, delimiter=',', unpack=True)
            ax.plot(nu1, C1)
            label = f.split('/')[1][:22]
            ax.set_title(label, fontsize=14)
            ax.text(0.1, 0.9, directory, fontsize=24, transform=ax.transAxes)
            ax.set_ylim(5000, 20000)
        plt.savefig('/content/' + directory + '_spectra.png', dpi=300)
        # plt.clf()
    return directory, nfiles
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os, glob
import csv
import scipy as sp
%matplotlib inline

# Function to examine all spectra in a directory and check for interference
# If any issue is identified, it is recommended to manually delete or rename the file before proceeding
def inspect(directory):
    files = glob.glob(directory + '/*.csv')
    nfiles = len(files)
    if nfiles == 0:
        print('No files found in directory ', directory)
        return 0
    else:
        fig, ax = plt.subplots(figsize=(12, 8))

        for n, f in enumerate(files):
            nu1, C1 = np.genfromtxt(f, delimiter=',', unpack=True)
            ax.plot(nu1, C1, label=f.split('/')[1][:22])

        ax.set_title(directory, fontsize=14)
        ax.set_xlabel('Frequency')
        ax.set_ylabel('Counts')
        ax.legend()
        plt.savefig('/content/' + directory + '_spectra.png', dpi=300)
    return directory, nfiles

# Change to the data directory
path = '/content/drive/MyDrive/Python/example_data'
os.chdir(path)
filelist = os.listdir()

# Iterate over the directories and perform the analysis of individual spectra
for f in filelist:
    print('Plotting ', f)
    inspect(f)
```
